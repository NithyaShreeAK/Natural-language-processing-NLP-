```python
# Step 1: Import Necessary Libraries
import spacy

# Step 2: Load the spaCy language model
nlp = spacy.load("en_core_web_sm")

# Step 3: Example sentences for different sentence types
interrogative_sentence = "What time does the train arrive?"
declarative_sentence = "The sun sets in the west."
complex_sentence = "Although it was raining, they continued their journey, and the sun appeared later."

# Step 4: Preprocess sentences
def preprocess(sentence):
    return nlp(sentence)

# Step 5: Function to analyze morphology
def analyze_morphology(doc):
    print("\nMorphological Analysis Results:\n")
    print("{:<15} {:<15} {:<10} {:<10} {:<30}".format("Word", "Lemma", "POS", "Tag", "Morphology"))
    print("=" * 75)
    for token in doc:
        print("{:<15} {:<15} {:<10} {:<10} {:<30}".format(
            token.text, token.lemma_, token.pos_, token.tag_, str(token.morph)
        ))

# Step 6: Analyze each sentence
print("\nInterrogative Sentence Analysis:")
interrogative_doc = preprocess(interrogative_sentence)
analyze_morphology(interrogative_doc)

print("\nDeclarative Sentence Analysis:")
declarative_doc = preprocess(declarative_sentence)
analyze_morphology(declarative_doc)

print("\nComplex Sentence Analysis:")
complex_doc = preprocess(complex_sentence)
analyze_morphology(complex_doc)

# Step 7: Evaluate Results (Note: Output will be in the terminal after running)
```

```python
# Step 1: Import the Necessary Libraries
import spacy
from spacy import displacy

# Step 2: Load the spaCy Language Model
nlp = spacy.load("en_core_web_sm")

# Step 3: Define the Sample Text
text = "The quick brown fox jumps over the lazy dog."

# Step 4: Process the Text
doc = nlp(text)

# Step 5: Print POS Tagging Results
print("Coarse-Grained and Fine-Grained POS Tagging Results:\n")
print(f"{'Token':10} {'Coarse-Grained (token.pos_)':30} {'Fine-Grained (token.tag_)':30}")
print("-" * 70)

# Step 6: Iterate Over Tokens and Display Their Tags
for token in doc:
    token_text = token.text
    coarse_tag = token.pos_  # Universal (coarse-grained) POS tag
    fine_tag = token.tag_  # Language-specific (fine-grained) POS tag
    print(f"{token_text:10} {coarse_tag:30} {fine_tag:30}")

# Step 7: Visualize the Dependency Parse Tree
displacy.render(doc, style="dep")
```
