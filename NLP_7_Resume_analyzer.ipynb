```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from google.colab import files

# Download NLTK data files (first time only)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab') # This line downloads the necessary punkt_tab data

# Define a function to clean and tokenize text
def clean_and_tokenize(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize text into words
    tokens = word_tokenize(text)
    # Remove stopwords (common words that don't add value, like 'and', 'the', etc.)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return filtered_tokens

# Function to score resumes based on keyword matching
def score_resume(resume_text, keywords):
    # Tokenize and clean the resume text
    resume_tokens = clean_and_tokenize(resume_text)

    # Create a counter for the tokens in the resume
    resume_counter = Counter(resume_tokens)

    # Calculate score based on how many keywords are present in the resume
    score = 0
    for keyword in keywords:
        score += resume_counter[keyword.lower()]

    return score

# Define your job description's relevant keywords
keywords = [
    "python", "data analysis", "machine learning", "sql", "statistics", "data visualization", "deep learning", "pandas", "numpy"
]

# Function to read uploaded resume files
def read_uploaded_resumes(uploaded_files):
    resumes = []
    for filename in uploaded_files:
        with open(filename, 'r', encoding='utf-8') as file:
            resumes.append((filename, file.read()))
    return resumes

# Example usage
if __name__ == "__main__":
    # Upload files
    uploaded = files.upload()

    # Read resumes from the uploaded files
    resumes = read_uploaded_resumes(uploaded.keys())

    # Score each resume based on keyword matches
    for resume_name, resume_text in resumes:
        score = score_resume(resume_text, keywords)
        print(f"Resume: {resume_name}, Score: {score}")
```

```python
import os
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter
from google.colab import files
from sklearn.feature_extraction.text import TfidfVectorizer

# Download necessary NLTK data files (first time only)
nltk.download('punkt')
nltk.download('stopwords')

# Define a function to clean and tokenize text
def clean_and_tokenize(text):
    # Convert to lowercase
    text = text.lower()
    # Tokenize text into words
    tokens = word_tokenize(text)
    # Remove stopwords (common words that don't add value, like 'and', 'the', etc.)
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
    return filtered_tokens

# Function to score resumes based on keyword matching
def score_resume(resume_text, keywords):
    # Tokenize and clean the resume text
    resume_tokens = clean_and_tokenize(resume_text)

    # Create a counter for the tokens in the resume
    resume_counter = Counter(resume_tokens)

    # Calculate score based on how many keywords are present in the resume
    score = 0
    for keyword in keywords:
        score += resume_counter[keyword.lower()]

    return score

# Define your job description's relevant keywords
keywords = [
    "python", "data analysis", "machine learning", "sql", "statistics", "data visualization", "deep learning", "pandas", "numpy"
]

# Function to read uploaded resume files
def read_uploaded_resumes(uploaded_files):
    resumes = []
    for filename in uploaded_files:
        with open(filename, 'r', encoding='utf-8') as file:
            resumes.append((filename, file.read()))
    return resumes

# Function to extract keywords using TF-IDF
def extract_keywords_using_tfidf(resumes):
    # Convert the list of resume texts into a list for TF-IDF processing
    resume_texts = [resume_text for _, resume_text in resumes]

    # Initialize TF-IDF vectorizer
    tfidf_vectorizer = TfidfVectorizer(stop_words='english')

    # Fit and transform the resume texts
    tfidf_matrix = tfidf_vectorizer.fit_transform(resume_texts)

    # Get feature names (words)
    feature_names = tfidf_vectorizer.get_feature_names_out()

    # Extract top keywords from each resume based on TF-IDF scores
    resume_keywords = {}

    for i, resume_text in enumerate(resume_texts):
        tfidf_scores = tfidf_matrix[i].toarray()[0]
        word_scores = list(zip(feature_names, tfidf_scores))

        # Sort the words by TF-IDF score (highest to lowest)
        sorted_word_scores = sorted(word_scores, key=lambda x: x[1], reverse=True)

        # Extract top 5 keywords for each resume
        resume_keywords[resumes[i][0]] = [word for word, score in sorted_word_scores[:5]]

    return resume_keywords

# Example usage
if __name__ == "__main__":
    # Upload files
    uploaded = files.upload()

    # Read resumes from the uploaded files
    resumes = read_uploaded_resumes(uploaded.keys())

    # Extract keywords using TF-IDF from the resumes
    resume_keywords = extract_keywords_using_tfidf(resumes)

    # Display the top keywords from each resume
    for resume_name, keywords in resume_keywords.items():
        print(f"Resume: {resume_name}")
        print(f"Extracted Keywords: {', '.join(keywords)}")
        print("-" * 50)

    # Score each resume based on keyword matches from your predefined list
    for resume_name, resume_text in resumes:
        score = score_resume(resume_text, keywords)
        print(f"Resume: {resume_name}, Match Score: {score}")
```
